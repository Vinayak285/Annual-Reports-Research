# Historical prices data

```python
import pickle
with open('nifty_next_tickers.pickle', 'rb') as f:
    tickers = pickle.load(f)
```

```python
yahoo_tickers = []
for ticker in tickers:
    ticker = f"{ticker}.NS"
    yahoo_tickers.append(ticker)
```

```python
start = '2011-03-01'
end = '2020-02-01'
```

```python
# pass yahoo_tickers in this func by appending '.NS' at the end of tickers' name
import os
import pandas_datareader as web
def get_data_from_yahoo(tickers, start, end):
    if not os.path.exists('prices'):
        os.makedirs('prices')

    for ticker in tickers:
        if not os.path.exists(f"prices/{ticker.split('.')[0]}.csv"):
            df = web.DataReader(ticker, 'yahoo', start, end)
            df.to_csv(f"prices/{ticker.split('.')[0]}.csv")
            print(f'Got {ticker}')
        else:
            print(f'Already have {ticker}')
```

```python
# pass yahoo ticker in this func as well
# keeping only the adj close as it adjusts for dividends and stock splits
import pandas as pd
import numpy as np

def compile_data(tickers): 
    main_df = pd.DataFrame()

    for count,ticker in enumerate(tickers):
            df = pd.read_csv(f"prices/{ticker.split('.')[0]}.csv")
            df.set_index('Date', inplace= True)
            name =ticker.split('.')[0]
            df.rename(columns = {'Adj Close': name}, inplace = True)
            df.drop(['Open', 'High', 'Low', 'Close', 'Volume'],1, inplace= True)

            if main_df.empty:
                    main_df = df
            else:
                    main_df = main_df.join(df, how = 'outer')

    main_df.replace('', np.nan)
    main_df.to_csv(f'concatenated_prices.csv') 
    
    return main_df
```

```python
get_data_from_yahoo(yahoo_tickers, start, end)
prices = compile_data(yahoo_tickers)
```
