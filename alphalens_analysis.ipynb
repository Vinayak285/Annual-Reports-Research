# This was the initial analysis that I did under the assumption that annual reports were available right after the closure of the financial year.

```python
import alphalens as al
from alphalens.performance import mean_information_coefficient
import pickle
import pandas as pd
```

```python
prices = pd.read_csv('concatenated_prices.csv', index_col = 0)
prices.index = pd.to_datetime(prices.index, utc = True, infer_datetime_format= True)
```

```python
factor_df = pd.read_csv('factor_data.csv', index_col = [0,1], parse_dates=True)
```

## For some reason, alphalens locally changes value of the periods 21 days become 34 days and so on, hence I have selected periods to be (1,3,15) to get (1,5,21).

## Individual factor analysis

```python
factors = ['Negative', 'Positive', 'Litigious', 'Uncertainty', 
           'StrongModal', 'WeakModal', 'Constraining',]

for sentiment in factors:
    
    sentiment_factor = al.utils.get_clean_factor_and_forward_returns(
                                                 factor = factor_df[sentiment],
                                                 prices = prices,
                                                 periods = (1,3,15),
                                                 quantiles = 5,
                                                 max_loss = 1);

    print(f"{sentiment} Tear Sheet")
    al.tears.create_summary_tear_sheet(sentiment_factor);
```

## Crude factor sentiment analysis

```python
crude_factor = al.utils.get_clean_factor_and_forward_returns(factor = factor_df['Crude_Sentiment'],
                                                       prices= prices,
                                                        periods = (1,3,15),
                                                      max_loss = 1);
al.tears.create_full_tear_sheet(crude_factor);
```

## MIx Sentiment factor analysis

```python
mix_factor = al.utils.get_clean_factor_and_forward_returns(factor = factor_df['Sentiment'],
                                                       prices= prices,
                                                      periods = (1,3,15),
                                                      max_loss = 1);
al.tears.create_full_tear_sheet(mix_factor);
```
